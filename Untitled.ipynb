# Análise de histórico das olimpiadas

## Introdução
Essa análise busca analisar um histórico a respeito dos atletas que participaram das olímpiadas desde 1900 e através dele encontrar padrões acerca dos medalhistas e conclusões interessantes em torno dos padrões neles.

### Setup

Algumas bibliotecas precisam ser importadas para realizar a análise. Isto feito, importa-se o csv usando o pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

historico = pd.read_csv('df_athlete_events.csv')
```

Com o dataset importado podemos dar uma olhada na cara dele:

```python
historico.head(5)
```

Realizar a analise de atleta por atleta pode se mostrar complexo, desta forma, é interessante destilar este dataset de forma a permitir a análise por comitê olímpico, isto será feito através da criação de um novo dataframe:

```python
comite=pd.DataFrame()
nocsRand=historico['NOC'].unique()
comite['NOC']=sorted(nocsRand)
gold = []
silver = []
bronze = []
none = []



for i in comite['NOC']:
    aux=historico[(historico['NOC']==i)&(historico['Medal']=='Gold')]
    gold.append(len(aux))
    aux=historico[(historico['NOC']==i)&(historico['Medal']=='Silver')]
    silver.append(len(aux))
    aux=historico[(historico['NOC']==i)&(historico['Medal']=='Bronze')]
    bronze.append(len(aux))
    aux=historico[(historico['NOC']==i)&(pd.isnull(historico['Medal']))]
    none.append(len(aux))
    
comite['Ouro'] = gold
comite['Prata'] = silver
comite['Bronze'] = bronze
comite['Nenhuma'] = none
comite
            
```

Note que este novo DF (dataframe) relaciona cada comitê com as medalhas adquiridas ao longo dos anos, desta forma é possível, por exemplo, averiguar quais são os 5 países com mais ouros na história das olimpíadas:

```python
comite.sort_values('Ouro', ascending=False).head(5)
```

```python
sns.barplot(y='Ouro', x='NOC', data=comite.sort_values('Ouro', ascending=False).head(5), palette='husl')
```

É interessante notar como os EUA ganham de forma esmagadora dos outros comitês em termos de medalhas de ouro, isso se dá não apenas pela competência dos atletas, mas também por se tratar de um dos poucos comitês olimpicos que manteve-se o mesmo ao longo de todo esse tempo; por exemplo, URS, relativo à finada União Soviética se divide em vários outros comitês, e GER, referente à alemanha, durante o período da guerra fria se divide em alemanha oriental e ocidental.

Porém não é justo analisar apenas a métrica de medalhas, uma vez que a quantidade de atletas e a quantidade de competições em que um comitê participou afetam drasticamente esse número cumulativo, desta forma, vamos adicionar mais algumas colunas à nossa tabela de comitês, colunas estas que serão relativas as participações, quantidade de atletas diferentes e modalidades:

```python
ano = []
atletas = []
modalidades = []

for i in comite['NOC']:
    ano.append(((historico[(historico['NOC']==i)]).nunique())['Year'])
    atletas.append(((historico[(historico['NOC']==i)]).nunique())['Name'])
    modalidades.append(((historico[(historico['NOC']==i)]).nunique())['Event'])

comite['Participações'] = ano
comite['Atletas individuais'] = atletas
comite['Modalidades'] = modalidades
comite
```

Outras colunas interessantes que podem ser extraidas das que já temos são as de total de medalhas, da relação de medalhas por atleta e das medalhas por participação, que por se tratarem de operações simples podem ser obtidas facilmente desta forma:

```python
comite['Total de medalhas'] = comite['Bronze']+comite['Prata']+comite['Ouro']
comite['Medalhas/Atleta'] = comite['Total de medalhas']/comite['Atletas individuais']
comite['Medalhas/Participação'] = comite['Total de medalhas']/comite['Participações']
comite
```

Com estas colunas em mãos, podemos analisar os resultados oriundos delas separadamente:

## Análise por sexo:

Antes de analisar o sexo dos atletas, é necessário fazer a limpeza de dados, uma vez que nenhum padrão foi adotado para o sexo no dataset, então ele pode assumir m/f, M/F ou MALE/FEMALE. Então para isso é possível corrigir eles rapidamente desta forma:

```python
sexTratado = []
for i in corrige['Sex']:
    if i == 'Male' or i=='M' or i=='m' or i=='MALE':
        sexTratado.append('M')
    elif i == 'Female' or i=='F' or i=='f' or i=='FEMALE':
        sexTratado.append('F')
    else:
        sexTratado.append(i)
sexTratado
corrige['Sex'] = sexTratado
# corrige[(corrige['Sex']!='F') & (corrige['Sex']!='M')]
corrige[(corrige['Sex']!='F') & (corrige['Sex']!='M')]
```

Mesmo assim, é possível notar que ainda possuimos cerca de 650 entradas cujo sexo está errado, porém o jogo (supostamente) está certo, então podemos usar isto para acertar a segunda coluna:

```python
def consertaSexo(linha):
    if linha['Sex'] != 'M' and linha['Sex'] != 'F':
        if 'Women' in linha['Event']:
            return 'F'
        elif 'Men' in linha['Event']:
            return 'M'
        else:
            return linha['Sex']
    else:
        return linha['Sex']
```

```python
teste = corrige
saida = []
saida = teste.apply(consertaSexo, axis=1)
teste['Sex'] = saida
corrige=teste
corrige[(corrige['Sex']!='F') & (corrige['Sex']!='M')].shape
```

Desta forma, podemos ver que as entradas restantes são apenas 35, o que é irrelevante pro tamanho desse dataset.
